```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('manager_data.xlsx')
df
```

```python
info_all=[]
for one_stock in df.groupby('Scode'):
    for one_year in one_stock[1].groupby('Date'):
#         print(one_year)
        for manager_group in one_year[1].groupby('if manager'):
            info=manager_group[1]['Ovesedu'].value_counts()
#             print(manager_group[0])
            if manager_group[0]==6:
                if 1 in info.index:
                    man_oversea=info.loc[1]
                else:
                    man_oversea=0
                if 0 in info.index:
                    man_native=info.loc[0]
                else:
                    man_native=0
            if manager_group[0]==0:
                if 1 in info.index:
                    non_man_oversea=info.loc[1]
                else:
                    non_man_oversea=0
                if 0 in info.index:
                    non_man_native=info.loc[0]
                else:
                    non_man_native=0
        one_list=manager_group[1].iloc[0,0:3].tolist()+[man_oversea,man_native,non_man_oversea,non_man_native]
        info_all.append(one_list)
```

```python
info_df=pd.DataFrame(info_all,columns=['code','name','date','man_oversea','man_native','non_man_oversea','non_man_native'])
info_df.to_excel('manager_info.xlsx')
```
